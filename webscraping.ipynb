```python

import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
import numpy as np
```

```python
request1=requests.get("https://www.flipkart.com/noise-colorfit-icon-2-1-8-display-bluetooth-calling-ai-voice-assistant-smartwatch/product-reviews/itmd4cd819eadb9f?pid=SMWGEH7VGYMGCP3V&lid=LSTSMWGEH7VGYMGCP3VXIJHDY&marketplace=FLIPKART")
request1.content
```

```python
soup=bs(request1.text,'html.parser')
soup
```

```python
review=soup.find_all("div",{"class":"ZmyHeo"})
reviews=[]
for re in review:
    retext=re.get_text().replace("READ MORE"," ").strip()
    reviews.append(retext)
    print(retext+"\n")   
```

```python
cust=soup.find_all("p",{"class":"_2NsDsF AwS1CA"})
customers=[]
for cu in cust:
    retext=cu.get_text()
    customers.append(retext)
    print(retext+"\n") 
```

```python
star=soup.find_all("div",{"class":"XQDdHH Ga3i8K"})
stars=[]
for st in star:
    retext=st.get_text()
    stars.append(retext)
    print(retext+"\n")
    
```

```python
tag=soup.find_all("p",{"class":"z9E0IG"})
tags=[]
for ta in tag:
    retext=ta.get_text()
    tags.append(retext)
    print(retext+"\n")
```

```python
minlen=min(len(reviews),len(stars),len(customers),len(tags))
print(minlen)
```

```python
data={
    "customers":customers[:minlen],
    "reviews":reviews[:minlen],
    "rating":stars[:minlen],
    "tags":tags[:minlen]
}
data
```

```python
df=pd.DataFrame(data)
df
```

```python
df.to_csv("review.csv")
```
