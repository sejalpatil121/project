```python
# Import necessary libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from scipy import stats
from scipy.stats import zscore

# Replace 'cleveland.csv' with your file path
df = pd.read_csv("heart.csv")
```

```python
print(df.isnull().sum())

for col in df:
  print(col, df[col].unique())
     
```

```python
# Replace '?' with NaN
df.replace('?', np.nan, inplace=True)

# Convert all columns to numeric where possible
df = df.apply(pd.to_numeric, errors='coerce')

# Drop rows with missing values
df.dropna(inplace=True)

# Remove rows with any negative values (if contextually invalid)
num_cols = df.select_dtypes(include=np.number).columns
for col in num_cols:
    df = df[df[col] >= 0]

print(df.isnull().sum())
df.shape

for col in df:
  print(col, df[col].unique())
```

```python
# Remove outliers using Z-score (threshold = 3)
z_scores = np.abs(stats.zscore(df[num_cols]))
df = df[(z_scores < 3).all(axis=1)]

print("After outlier removal:", df.shape)
```

```python
# Separate features and target
X = df.drop('target', axis=1)  # Replace 'target' with actual target column if different
y = df['target']

# Normalize numeric features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
# Logistic Regression
lr = LogisticRegression()
lr.fit(X_train, y_train)
```

```python
lr_pred = lr.predict(X_test)
lr_acc = accuracy_score(y_test, lr_pred) * 100

print(f"Logistic Regression Accuracy: {lr_acc:.2f}%")
```

```python

# k-NN
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
```

```python
knn_pred = knn.predict(X_test)
knn_acc = accuracy_score(y_test, knn_pred) * 100

print(f"k-NN Accuracy: {knn_acc:.2f}%")
```
