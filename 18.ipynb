```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('heart.csv')

# Replace '?' with NaN and drop rows with missing values
df.replace('?', pd.NA, inplace=True)
df.dropna(inplace=True)

# Convert data types
df = df.astype({
    'ca': 'float64',
    'thal': 'float64'
})

```

```python
plt.figure(figsize=(10, 6))
df['age'].hist(bins=20, color='skyblue', edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.grid(False)
plt.show()
```

```python
labels = ['No Disease', 'Heart Disease']
sizes = df['target'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=['lightgreen', 'salmon'])
plt.title('Proportion of Heart Disease Cases')
plt.axis('equal')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='target', y='chol', hue='target', data=df, palette='Set2')
plt.title('Cholesterol Level by Heart Disease')
plt.xlabel('Heart Disease (1=Yes, 0=No)')
plt.ylabel('Cholesterol')
plt.legend([], [], frameon=False)  # Hides legend if not needed
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='age', y='thalach', hue='target', data=df, palette='coolwarm')
plt.title('Age vs Max Heart Rate by Heart Disease')
plt.xlabel('Age')
plt.ylabel('Max Heart Rate')
plt.show()
```

```python
import seaborn.objects as so  # Seaborn v0.12+ (optional) or use `sns.jointplot` alternative

g = sns.jointplot(data=df, x='age', y='chol', hue='target', kind='scatter', palette='Set1')
g.plot_marginals(sns.boxplot, color='lightgray')  # Removed alpha
g.fig.suptitle('Scatter Plot of Age vs Cholesterol with Boxplots', y=1.03)
plt.show()

```
