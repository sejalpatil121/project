```python
import pandas as pd
import numpy as np

# Load the dataset
df = pd.read_csv('hepatitis_csv.csv')

print(df.isnull().sum())
```

```python
df.replace('?', np.nan, inplace=True)
df.dropna(inplace=True)# b. Merge two subsets (concat vertically)

#df = df.applymap(lambda x: x.strip() if isinstance(x, str) else x)
for col in df.columns:
    try:
        df[col] = pd.to_numeric(df[col], errors='coerce')  # Convert non-numeric columns to NaN
    except:
        pass

print(df.isnull().sum())
```

```python
# a. Create data subsets for different sex
#male_df = df[df['Sex'] == 1]
#female_df = df[df['Sex'] == 2]

# a. Create data subsets for different sex using loc[]
male_df = df.loc[df['Sex'] == 1]
female_df = df.loc[df['Sex'] == 2]
female_df
```

```python
# b. Merge two subsets (concat vertically)
merged_df = pd.concat([male_df, female_df], axis=0)
merged_df
```

```python
# c. Sort Data using Age, SGOT, PROTIME
sorted_df = merged_df.sort_values(by=['Age', 'SGOT', 'Protime'])
sorted_df
```

```python
# d. Transposing Data (first 5 rows for clarity)
transposed_df = sorted_df.head().transpose()
transposed_df
```

```python
# e. Melting Data to long format (example with selected columns)
melted_df = pd.melt(
    df[['Age', 'Sex', 'SGOT', 'Protime']],
    id_vars=['Sex'],
    var_name='Variable',
    value_name='Value'
)
melted_df
```

```python
# f. Casting (pivoting) data to wide format
casted_df = melted_df.pivot_table(index='Sex', columns='Variable', values='Value', aggfunc='mean').reset_index()
casted_df
```

```python
# Display outputs
print("\n--- Male Subset ---\n", male_df.head())
print("\n--- Female Subset ---\n", female_df.head())
print("\n--- Merged Data ---\n", merged_df.head())
print("\n--- Sorted Data ---\n", sorted_df[['Age', 'SGOT', 'Protime']].head())
print("\n--- Transposed Data (First 5 rows) ---\n", transposed_df)
print("\n--- Melted Data ---\n", melted_df.head())
print("\n--- Casted Data (Wide Format) ---\n", casted_df)
```
