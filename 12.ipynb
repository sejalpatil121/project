```python
import pandas as pd

# Load dataset with no header and assign column names
df = pd.read_csv("adult_dataset.csv")

# Optional: print first few rows to confirm
print(df.head())
print(df.isnull().sum())
```

```python
df = df.applymap(lambda x: x.strip() if isinstance(x, str) else x)
#subset_country = df[df['native-country'] == 'United-States']
#subset_sex = df[df['sex'] == 'Male']
#subset_race = df[df['race'] == 'White']

subset_country = df.loc[df['native-country'] == 'United-States']
subset_sex = df.loc[df['sex'] == 'Male']
subset_race = df.loc[df['race'] == 'White']
```

```python
merged_data = pd.concat([subset_country, subset_race], ignore_index=True)
merged_data
```

```python
melted = pd.melt(df, id_vars=['sex'], value_vars=['age', 'educational-num', 'hours-per-week'],
                 var_name='Attribute', value_name='Value')
melted
```
