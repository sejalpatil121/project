```python
import pandas as pd

# Load dataset
df = pd.read_csv('iris.csv')

print(df.isnull().sum())
df.head()
```

```python
# Create subsets for each species
#setosa_df = df[df['species'] == 'Iris-setosa']
#versicolor_df = df[df['species'] == 'Iris-versicolor']
#virginica_df = df[df['species'] == 'Iris-virginica']

# Create subsets using .loc[]
setosa_df = df.loc[df['variety'] == 'Setosa']
versicolor_df = df.loc[df['variety'] == 'Versicolor']
virginica_df = df.loc[df['variety'] == 'Virginica']
```

```python
merged_df = pd.concat([setosa_df, versicolor_df])
merged_df
```

```python
sorted_df = df.sort_values(by='petal.length', ascending=True)
sorted_df
```

```python
transposed_df = df.head(5).T  # Transposing only top 5 rows for visibility
transposed_df
```

```python
# Melting feature columns
melted_df = pd.melt(df,
                    id_vars=['variety'],
                    value_vars=['sepal.length', 'sepal.width', 'petal.length', 'petal.width'],
                    var_name='feature',
                    value_name='value')
# Display melted DataFrame
print(melted_df.head())
```

```python
wide_df = melted_df.pivot_table(index='variety', columns='feature', values='value', aggfunc='mean')
print(wide_df.head())
wide_df.reset_index(inplace=True)
```
