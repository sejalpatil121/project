```python
import pandas as pd

# Load dataset
df = pd.read_csv('dataset_Facebook.csv', sep=';')
print(df.isnull().sum())
```

```python
df.fillna({
    'Paid': df['Paid'].mode()[0],
    'like': df['like'].mode()[0],
    'share': df['share'].mode()[0]
}, inplace=True)
print(df.isnull().sum())
print(df.columns.tolist())
df.shape
df['Type'].unique()
```

```python
# Subsets by type of post using loc[]
#loc[] is label-based indexing and is best when you want to filter based on column values.
photo_posts = df.loc[df['Type'] == 'Photo']
video_posts = df.loc[df['Type'] == 'Video']
status_posts = df.loc[df['Type'] == 'Status']
link_posts = df.loc[df['Type'] == 'Link']
# Create subsets for each species
#setosa_df = df[df['species'] == 'Iris-setosa']

#which is index-based selection (purely integer positions).
#subset_rows = df.iloc[0:5, 2:4]
#subset_rows

# First subset: Like and Share
#df_subset_1 = df[['like','share']]
#df_subset_1
```

```python
# Merge photo and video data (vertical concatenation)
merged_df = pd.concat([photo_posts, video_posts], ignore_index=True)
merged_df
```

```python
# Sort in descending order
sorted_df = df.sort_values(by='Page total likes', ascending=True)
sorted_df
```

```python
# Transpose the first 5 rows for readability
transposed_df = merged_df.transpose()
transposed_df.head()
```

```python
# Example with 3 columns: Lifetime Post Total Reach and Lifetime Engaged Users
melted_df = pd.melt(df, 
                    id_vars=['Type', 'Post Month'], 
                    value_vars=['Lifetime Post Total Reach', 'Lifetime Engaged Users'],
                    var_name='Metric', 
                    value_name='Value')
melted_df
```

```python
# Pivot back to wide format
wide_df = melted_df.pivot_table(index=['Type', 'Post Month'], 
                                 columns='Metric', 
                                 values='Value', 
                                 aggfunc='sum').reset_index()
wide_df
```
