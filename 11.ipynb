```python
import pandas as pd

# Load the movie dataset
df = pd.read_csv("movies_metadata.csv")  
```

```python
df.isnull().sum()
```

```python
for col in df.columns:
    if df[col].isnull().sum() > 0:
        df[col] = df[col].fillna(df[col].mode()[0])
```

```python
df.isnull().sum()
```

```python
# Subset for English movies using loc[]
english_movies = df.loc[df['original_language'] == 'en']
# Subset for Hindi movies using loc[]
hindi_movies = df.loc[df['original_language'] == 'hi']
# Subset for French movies using loc[]
french_movies = df.loc[df['original_language'] == 'fr']

# Subset for English language
#english_movies = df[df['original_language'] == 'en']
# Subset for Hindi language
#hindi_movies = df[df['original_language'] == 'hi']
```

```python
# Merge English and Hindi movies
merged_movies = pd.concat([english_movies, hindi_movies], ignore_index=True)
merged_movies
```

```python
# Sort by vote_average in descending order
sorted_movies = df.sort_values(by='vote_average', ascending=False)
print(sorted_movies.head())
```

```python
# Transpose a small sample of data for better readability
transposed = df.head(2).transpose()
transposed
```

```python
# Melting
melted_df = pd.melt(df,
                    id_vars=['original_title'],
                    value_vars=['budget', 'popularity'],
                    var_name='feature',
                    value_name='value')
melted_df
```

```python
casted_df = melted_df.pivot_table(
    index='original_title',
    columns='feature',
    values='value',
    aggfunc='first'   # or 'mean', 'max', etc.
).reset_index()
casted_df
```
