```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('adult_dataset.csv')


# Clean column names (remove leading/trailing spaces)
df.columns = [col.strip() for col in df.columns]

# Preview the dataset
print(df.head())
```

```python
# Histogram of Age
plt.figure(figsize=(8, 5))
plt.hist(df['age'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Pie chart for income
income_counts = df['income'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(income_counts, labels=income_counts.index, autopct='%1.1f%%', startangle=140, colors=['#66b3ff', '#ff9999'])
plt.title('Income Distribution')
plt.axis('equal')
plt.show()
```

```python
# Box plot of age by income
plt.figure(figsize=(8, 5))
sns.boxplot(x='income', y='age', data=df)
plt.title('Age Distribution by Income')
plt.show()
```

```python
# Scatter plot of age vs hours-per-week
plt.figure(figsize=(8, 6))
sns.scatterplot(x='age', y='hours-per-week', hue='income', data=df, alpha=0.6)
plt.title('Age vs Hours-per-Week by Income')
plt.show()
```

```python
# Jointplot with marginal boxplots
g = sns.jointplot(
    data=df,
    x="age", y="hours-per-week",
    hue="income",
    kind="scatter",
    marginal_ticks=True,
    marginal_kws=dict(common_norm=False)
)

# Add boxplots manually (use `palette` instead of `color` for gradient)
g.plot_marginals(sns.boxplot, palette="dark:lightgray")  # Use palette instead of color
plt.show()
```
